```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
state_list1 = np.load('./experiment data/state_list_2500.npy')
state_list2 = np.load('./experiment data/state_list_5000.npy')
state_list3 = np.load('./experiment data/state_list_7500.npy')
state_list4 = np.load('./experiment data/state_list_10000.npy')
```

```python
indexes = np.where(~np.isnan(state_list1[0,:,0,0]))[0][-1] + 1
plt.plot((state_list1[0,:indexes,0,0]==2).cumsum())

indexes = np.where(~np.isnan(state_list2[0,:,0,0]))[0][-1] + 1
plt.plot((state_list2[0,:indexes,0,0]==2).cumsum())

indexes = np.where(~np.isnan(state_list3[0,:,0,0]))[0][-1] + 1
plt.plot((state_list3[0,:indexes,0,0]==2).cumsum())

indexes = np.where(~np.isnan(state_list4[0,:,0,0]))[0][-1] + 1
plt.plot((state_list4[0,:indexes,0,0]==2).cumsum())

plt.legend(labels=['buyer SR model(penalty = 2500)',
                   'buyer SR model(penalty = 5000)',
                   'buyer SR model(penalty = 7500)',
                   'buyer SR model(penalty = 10000)'],loc = 3,bbox_to_anchor = (1,0))
plt.title('buyer telling truth for SR model')
```

```python
indexes = np.where(~np.isnan(state_list1[0,:,1,1]))[0][-1] + 1
plt.plot((state_list1[0,:indexes,1,1]==2).cumsum())

indexes = np.where(~np.isnan(state_list2[0,:,1,1]))[0][-1] + 1
plt.plot((state_list2[0,:indexes,1,1]==2).cumsum())

indexes = np.where(~np.isnan(state_list3[0,:,1,1]))[0][-1] + 1
plt.plot((state_list3[0,:indexes,1,1]==2).cumsum())

indexes = np.where(~np.isnan(state_list4[0,:,1,1]))[0][-1] + 1
plt.plot((state_list4[0,:indexes,1,1]==2).cumsum())
```
