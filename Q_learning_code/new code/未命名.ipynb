```python
indexes = np.where(~np.isnan(state_list2[0,:,0,0]))[0][-1] + 1
```

```python
indexes
```

```python
state_list2 = state_list2[0,:indexes,:,:]
```

```python
agree_ab_truth_list = np.full((indexes),np.nan)
disagree_ab_truth_list = np.full((indexes),np.nan)
agree_ab_lies_list = np.full((indexes),np.nan)
disagree_ab_lies_list = np.full((indexes),np.nan)
for i in range(len(state_list2)):
    if i%1000 ==0:
        print(i)
    if i>=1000:
        buyer_list = state_list2[i-1000:i,0,:]
        seller_list = state_list2[i-1000:i,1,:]
        
        agree_ab_truth_list[i-1000] = 0
        disagree_ab_truth_list[i-1000] = 0
        agree_ab_lies_list[i-1000] = 0
        disagree_ab_lies_list[i-1000] = 0
        for j in range(1000):
            vb, cb, vb_a = buyer_list[j]
            vs, cs, cs_a = seller_list[j]
            if vb != vs or cb != cs:
                if vb_a == 2 or cs_a == 2:                                
                    disagree_ab_truth_list[i-1000]+=1
                else:
                    disagree_ab_lies_list[i-1000]+=1
            if vb == vs and cb == cs:
                if cb == 2 or vs == 2:                                      
                    agree_ab_truth_list[i-1000]+=1
                else:
                    agree_ab_lies_list[i-1000]+=1  
```

```python
plt.plot(agree_ab_lies_list/1000)
plt.plot(agree_ab_truth_list/1000)
plt.plot(disagree_ab_lies_list/1000)
plt.plot(disagree_ab_truth_list/1000)
plt.legend(labels=['agree_on_lies_list','agree_on_truth_list',\
                  'disagree_arbitration_lies_list','disagree_abitration_truth_list'],loc='best')
plt.xlabel('converge_times')
plt.ylabel('ratio')
plt.title('ratio of disagree and agree and telling truth on penalty10000')
```

```python
plt.plot(disagree_ab_lies_list/(disagree_ab_lies_list+disagree_ab_truth_list))
plt.plot(disagree_ab_truth_list/(disagree_ab_lies_list+disagree_ab_truth_list))
plt.legend(labels=['disagree_arbitration_lies_list','disagree_abitration_truth_list'],loc='best')
plt.xlabel('converge_times')
plt.ylabel('ratio')
plt.title('ratio of disagree on telling truth or not on penalty7500')
```
